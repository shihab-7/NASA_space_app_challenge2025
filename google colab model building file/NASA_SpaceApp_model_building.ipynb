Night light clean and processed data

```python
import pandas as pd
import json
import numpy as np

ntl_df = pd.read_csv('/content/drive/MyDrive/NASA SpaceApp/Bangladesh_Division_Night_Light_Intensity_2017_2024.csv')

columns_to_drop =['ADM0_CODE','ADM1_CODE','DISP_AREA','EXP1_YEAR','STATUS','STR1_YEAR','system:index']
ntl_df = ntl_df.drop(columns=columns_to_drop)

# Function to extract 'type' from JSON string
def extract_geo_type(geo_json):
    if pd.isna(geo_json):
        return np.nan
    try:
        geo_data = json.loads(geo_json)
        return geo_data.get('type', np.nan)
    except json.JSONDecodeError:
        return np.nan

# Apply the function to the '.geo' column
ntl_df['.geo'] = ntl_df['.geo'].apply(extract_geo_type)
ntl_df = ntl_df.rename(columns={'mean': 'night_light_mean'})

display(ntl_df)
```

population density actual csv

```python
import pandas as pd
import json
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/NASA SpaceApp/Bangladesh_Division_PopDensity_2017_2020.csv')

columns_to_drop =['ADM0_CODE','ADM1_CODE','DISP_AREA','EXP1_YEAR','STATUS','STR1_YEAR','system:index']
df = df.drop(columns=columns_to_drop)

def extract_geo_type(geo_json):
    if pd.isna(geo_json):
        return np.nan
    try:
        geo_data = json.loads(geo_json)
        return geo_data.get('type', np.nan)
    except json.JSONDecodeError:
        return np.nan
df['.geo'] = df['.geo'].apply(extract_geo_type)
df = df.rename(columns={'mean': 'popu_dense_mean'})

display(df)
```

population density scaling upto 2024 using regression

```python
# Group data by division and year and get the mean population density
pop_density_ts = df.groupby(['ADM1_NAME', 'Year'])['popu_dense_mean'].mean().unstack()

display(pop_density_ts)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd

forecasted_data = pd.DataFrame(columns=['ADM1_NAME', 'Year', 'popu_dense_mean'])

future_years = np.array([2021, 2022, 2023, 2024]).reshape(-1, 1)

for division in pop_density_ts.index:
    historical_data = pop_density_ts.loc[division].dropna()
    years = historical_data.index.values.reshape(-1, 1)
    density = historical_data.values

    # Fit a linear regression model
    model = LinearRegression()
    model.fit(years, density)

    # Predict future population density
    future_density = model.predict(future_years)

    # Create a temporary DataFrame for forecasted values
    temp_df = pd.DataFrame({
        'ADM1_NAME': division,
        'Year': future_years.flatten(),
        'popu_dense_mean': future_density
    })

    # Concatenate with the main forecasted_data DataFrame
    forecasted_data = pd.concat([forecasted_data, temp_df], ignore_index=True)

# Combine historical and forecasted data
scaled_pop_density = pd.concat([df[['ADM1_NAME', 'Year', 'popu_dense_mean']], forecasted_data], ignore_index=True)

scaled_pop_density = scaled_pop_density.sort_values(by=['Year']).reset_index(drop=True)

display(scaled_pop_density)
```

```python
from sklearn.linear_model import LinearRegression

forecasted_data = pd.DataFrame(columns=['ADM1_NAME', 'Year', 'popu_dense_mean'])

future_years = np.array([2021, 2022, 2023, 2024]).reshape(-1, 1)

for division in pop_density_ts.index:
    historical_data = pop_density_ts.loc[division].dropna()
    years = historical_data.index.values.reshape(-1, 1)
    density = historical_data.values

    model = LinearRegression()
    model.fit(years, density)

    future_density = model.predict(future_years)

    temp_df = pd.DataFrame({
        'ADM1_NAME': division,
        'Year': future_years.flatten(),
        'popu_dense_mean': future_density
    })

    forecasted_data = pd.concat([forecasted_data, temp_df], ignore_index=True)

scaled_pop_density = pd.concat([df[['ADM1_NAME', 'Year', 'popu_dense_mean']], forecasted_data], ignore_index=True)

# Sort by year for better visualization
scaled_pop_density = scaled_pop_density.sort_values(by=['ADM1_NAME', 'Year']).reset_index(drop=True)

display(scaled_pop_density)
```

```python
merged_df = pd.merge(df, scaled_pop_density[['ADM1_NAME', 'Year', 'popu_dense_mean']],
                     on=['ADM1_NAME', 'Year'],
                     how='right',
                     suffixes=('_original', '_scaled'))

merged_df['popu_dense_mean'] = merged_df['popu_dense_mean_scaled']
merged_df = merged_df.drop(columns=['popu_dense_mean_original', 'popu_dense_mean_scaled'])


original_cols = df.drop(columns=['Year', 'popu_dense_mean']).drop_duplicates(subset=['ADM1_NAME'])

for index, row in original_cols.iterrows():
    division_name = row['ADM1_NAME']
    missing_rows_mask = (merged_df['ADM1_NAME'] == division_name) & (merged_df['ADM0_NAME'].isnull())

    for col in original_cols.columns:
        if col != 'ADM1_NAME':
             merged_df.loc[missing_rows_mask, col] = row[col]


# Sort by division and year for better viewing
merged_df = merged_df.sort_values(by=['Year']).reset_index(drop=True)
pop_density_diff = merged_df

display(pop_density_diff)
```

multi-model pipeline with randomforest regressor and random forest classifier

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/NASA SpaceApp/combined_data.csv")

X = df[['Year', 'ADM1_NAME', 'Shape_Area', 'Shape_Leng']]
y = df[['night_light_mean', 'ndvi_mean', 'popu_dense_mean']]

# Encode categorical ADM1_NAME
X = pd.get_dummies(X, columns=['ADM1_NAME'])


reg = MultiOutputRegressor(RandomForestRegressor(n_estimators=200, random_state=42))
reg.fit(X, y)

y_pred = reg.predict(X)

composite = (df['night_light_mean'] + df['popu_dense_mean'] - df['ndvi_mean'])
df['class_label'] = pd.qcut(composite, 4, labels=['low', 'moderate', 'high','severe'])

clf = RandomForestClassifier(n_estimators=200, random_state=42)
clf.fit(y, df['class_label'])

# Predict final classes based on regressor outputs
final_classes = clf.predict(y_pred)

print(final_classes[:10])
```

```python
# manual_input = pd.DataFrame({
#     'Year': [2020],
#     'ADM1_NAME': ['Dhaka'],
#     'Shape_Area': [2.748844],
#     'Shape_Leng': [12.226167]
# })

# Let's generate a few random inputs based on the training data's characteristics
num_random_inputs = 5
random_inputs = []

for _ in range(num_random_inputs):
    random_year = np.random.randint(df['Year'].min(), df['Year'].max() + 5) # Random year within historical range + 5 years
    random_division = np.random.choice(df['ADM1_NAME'].unique())
    # For Shape_Area and Shape_Leng, we can sample from existing values or use representative values
    # Let's sample from existing values for simplicity
    random_shape_area = np.random.choice(df['Shape_Area'])
    random_shape_leng = np.random.choice(df['Shape_Leng'])

    random_inputs.append({
        'Year': random_year,
        'ADM1_NAME': random_division,
        'Shape_Area': random_shape_area,
        'Shape_Leng': random_shape_leng
    })

manual_input = pd.DataFrame(random_inputs)

# Use the same categories as used during training
manual_input_encoded = pd.get_dummies(manual_input, columns=['ADM1_NAME'])

# Ensure the columns in the manual input match the columns used during training (X)
# Reindex the manual input DataFrame to match the columns of X used during training, filling missing columns with 0
# This is important to ensure the order and presence of one-hot encoded columns are consistent
# Assuming X is available from the previous training cell
missing_cols = set(X.columns) - set(manual_input_encoded.columns)
for c in missing_cols:
    manual_input_encoded[c] = 0
manual_input_encoded = manual_input_encoded[X.columns]

predicted_means = reg.predict(manual_input_encoded)

predicted_means_df = pd.DataFrame(predicted_means, columns=y.columns)

predicted_class = clf.predict(predicted_means_df)

print("Random Inputs and Predicted Outputs:")
for i in range(num_random_inputs):
    print(f"\nInput {i+1}:")
    print(manual_input.iloc[i])
    print("\nPredicted mean values (Night Light, NDVI, Population Density):")
    print(predicted_means_df.iloc[[i]])
    print("\nPredicted class:")
    print(predicted_class[i])
```
